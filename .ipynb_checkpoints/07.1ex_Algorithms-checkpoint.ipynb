1\. **PCA on 3D dataset**

* Generate a dataset with 3 features each with N entries (N being ${\cal O}(1000)$). With $N(\mu,\sigma)$ the normali distribution with mean $\mu$ and $\sigma$  standard deviation, generate the 3 variables $x_{1,2,3}$ such that:
    * $x_1$ is distributed as $N(0,1)$
    * $x_2$ is distributed as $x_1+N(0,3)$
    * $x_3$ is given by $2x_1+x_2$
* Find the eigenvectors and eigenvalues of the covariance matrix of the dataset
* Find the eigenvectors and eigenvalues using SVD. Check that the two procedures yield to same result
* What percent of the total dataset's variability is explained by the principal components? Given how the dataset was constructed, do these make sense? Reduce the dimensionality of the system so that at least 99% of the total variability is retained.
* Redefine the data in the basis yielded by the PCA procedure
* Plot the data points in the original and the new coordiantes as a set of scatter plots. Your final figure should have 2 rows of 3 plots each, where the columns show the (0,1), (0,2) and (1,2) proejctions.


```python
#dataset 3 features w N rows 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
# column 1 
point1 = abs(np.random.normal(0, 1, 100)) 
# column 2 
point2 = abs(np.random.normal(0, 3, 100) + point1) 
# column 3 
point3 = 2*point1 + point2
x = np.c_[point1, point2, point3]

# defining a pandas data frame 
data = pd.DataFrame() 
  
# defining the columns of the dataset 
data['x1'] = point1 
data['x2'] = point2 
data['x3'] = point3
```

```python
data
```

```python
#Find the eigenvectors and eigenvalues of the covariance matrix of the dataset
np.cov(data) #covariance matrix
```

```python
from scipy import linalg as la
from matplotlib import pyplot as plt
A = data
```

```python
l, V = la.eig(np.cov(A))
print((l)) #l are eigenvalues
print(np.real(V)) #eigenvectors are columns of V matrix

b = V[0:100,0:1] * np.cov(A)
print(np.real(b))
b == np.cov(A)
```

```python
#Find the eigenvectors and eigenvalues using SVD. Check that the two procedures yield to same result
U, D, Vt = la.svd(A)
print (U,'\n') #eigenvectors
#eigenvalues D2
print((D * D)/99)
print(D.shape)
#print (Vt,'\n')
np.allclose(A, np.dot(U[:, :3] * D, Vt))
print(np.dot(U[:, :3] * D, Vt))
print(A)
```

```python
#print(l)
print((D*D)/99)
l_svd = D*D/99
V_svd = U 
```

```python
#What percent of the total dataset's variability is explained by the principal components? 
#Given how the dataset was constructed, do these make sense? 
#Reduce the dimensionality of the system so that at least 99% of the total variability is retained.

l, V = np.linalg.eig(np.cov(A))
Lambda=np.diag(l)
print (Lambda)
print ("A.trace():", np.cov(A).trace())
print ("Lambda.trace():", Lambda.trace())
#print (Lambda[0:5,0:5])
percentage = (Lambda[0,0]/Lambda.trace())* 100
percentage = round(percentage, 2)
print ("percentage variability retained", np.real(percentage),"%")

```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
principleComponents = pca.fit_transform(A)
principleDf = pd.DataFrame(data = principleComponents
             , columns = ['principle component 1', 'principle component 2'])
print(principleDf)
#print(A)
b = pca.explained_variance_ratio_
print(b)
print("percentage variability retained", b[0]+b[1])
```
